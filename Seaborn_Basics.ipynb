```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
df = pd.read_csv('https://raw.githubusercontent.com/DeanPhillipsOKC/machine-learning-samle-data/master/heart.csv')
df
```

## importing Seaborn

```
import seaborn as sns
```

# Seaborn Basics

## Creating a distribution plot

```
sns.distplot(df['age'])
```

The command above produces a plot with a KDE (Kernel Desity Estimation) which is a visualization of what a probability function would look like, and a histogram.

The KDE can be turned off by passing a false in for the kde parameter

```
sns.distplot(df['age'], kde=False)
```

### Bins
Each bar on the histogram represents a bin which contains the count of x axis elements that fall within a range.  In the example above we have approximiately 60 records that belong to patience whos age is in a range of about 57 to 60 years old.

We can change the number of bins using the bins parameter

```
sns.distplot(df['age'], kde=False, bins=3)
```

### Changing Size
You can control the size of the rendered image by calling Matplotlib directly before you call Seaborn

```
plt.figure(figsize=(10, 5))
sns.distplot(df['age'])
```

```
plt.figure(figsize=(5, 3))
```

### Changing the color, and adding some more bins

```
sns.distplot(df['age'], color='red', bins=50)
```

## Creating Count Plots
Count plots will let us partition data by features

```
sns.countplot(x='sex', data=df)
```

### Using hues to add subdivisions to the count plot partitions
The plot below creates a partition for each type of chest pain.  Each of those partitions are further subdivided by sex.

```
sns.countplot(x='cp', data=df, hue='sex')
```

## Creating Box Plots

With box plots your x feature should be catogorical whereas your y should be continuous data

```
sns.boxplot(x='sex', y='age', data=df)
```

In this case sex 1 appears to be a little younger on average than sex 0 but they are close

## Analyze box plot data further by using a hue
In this case we are using maximum heartbeat rate achieved as our continuos feature (i.e. y value), and target as our categorical feature

Our 'target' for this study is people with heart disease.

So ultimately we are comparing the max heartbeat achieved for folks with heart disease vs those without.

```
sns.boxplot(x='target', y='thalach', data=df)
```

In this case we have a much bigger discrepency, so lets add a hue based on sex to examine the impact

```
sns.boxplot(x='target', y='thalach', data=df, hue='sex')
```

## Scatter plots

### Using a hue
The first example uses a hue based on sex to add another category for analysis.

```
sns.scatterplot(x='chol', y='trestbps', data=df, hue='sex')
```

### Using Size
You can add another variable to the visualization by using the size attribute.  In this case points on the plot will be larger, the older the person is

```
plt.figure(figsize=(10, 7))
sns.scatterplot(x='chol', y='trestbps', data=df, hue='sex', size='age')
```

## Pair Plots


```
iris = pd.read_csv('https://raw.githubusercontent.com/DeanPhillipsOKC/machine-learning-samle-data/master/iris.csv')
iris
```

```
sns.pairplot(iris)
```

Each feature is laid out in both the x and y axis of the pair plot

Where the x and y features match (along the diagonal) is a histogram for that feature

everywhere else, there is a scatterplot of the x and y feature

### Adding Hues
this can help us get a better understanding of the data by adding another dimension to the analysis

One thing noteworth is that the histograms have been replaced with KDEs (Kernel Density Estimations)

```
sns.pairplot(iris, hue='species')
```
